<h2 class="western"><a name="__RefHeading___Toc253_3855830288"></a> 2.2 Step
2: Assessing data&nbsp;</font></font></h2>
<p class="western" align="left" style="font-variant: normal; letter-spacing: normal; font-style: normal; font-weight: normal">
<font color="#000000"><font face="FreeSans, sans-serif"><font size="4" style="font-size: 14pt">In
his paper Hadley Wickham sets the standard for tackling data tidiness
prior to data quality. In the abstract of the paper Wickham explains
that: &quot;Tidy datasets are easy to manipulate, model and
visualize, and have a specific structure: each variable is a column,
each observation is a row, and each type of observational unit is a
table.&quot;</font></font></font></p>
<p class="western" align="left" style="font-variant: normal; letter-spacing: normal; font-style: normal; font-weight: normal">
<font color="#000000"><font face="Helvetica Neue, Helvetica, Arial, sans-serif"><font size="2" style="font-size: 10pt"><font face="FreeSans, sans-serif"><font size="4" style="font-size: 14pt">Reference
:&nbsp;</font></font><a class="western" href="https://www.jstatsoft.org/article/view/v059i10/"><font color="#337ab7"><font face="FreeSans, sans-serif"><font size="4" style="font-size: 14pt"><u><span style="background: #ffffff">https://www.jstatsoft.org/article/view/v059i10/</span></u></font></font></font></a></font></font></font></p>
<p class="western" align="left" style="font-variant: normal; letter-spacing: normal; font-style: normal; font-weight: normal; text-decoration: none">
<font color="#000000"><font face="FreeSans, sans-serif"><font size="4" style="font-size: 14pt"><span style="background: #ffffff">Considering
the above abstract from Wickham, the gathered data made up of the
three collected data files through the three different modes were
tackled accordingly.</span></font></font></font></p>
<p class="western" align="left" style="font-variant: normal; letter-spacing: normal; font-style: normal; font-weight: normal; text-decoration: none">
<font color="#000000"><font face="FreeSans, sans-serif"><font size="4" style="font-size: 14pt"><span style="background: #ffffff">This
approach resulted in identifying two tidiness issues&nbsp;:</span></font></font></font></p>
<ul>
	<li><p class="western" align="left" style="font-variant: normal; letter-spacing: normal; font-style: normal; font-weight: normal; text-decoration: none">
	<font color="#000000"><font face="FreeSans, sans-serif"><font size="4" style="font-size: 14pt"><span style="background: #ffffff">The
	data was all part of the same observation and the files are required
	to be combined into a single data frame to make the observational
	unit with the above referenced recommendations</span></font></font></font></p>
	<li><p class="western" align="left" style="font-variant: normal; letter-spacing: normal; font-style: normal; font-weight: normal; text-decoration: none">
	<font color="#000000"><font face="FreeSans, sans-serif"><font size="4" style="font-size: 14pt"><span style="background: #ffffff">Inline
	with having every observation as a row, it was noted that the
	dogtionary ratings were split into columns and required the change
	in rows per tweet id</span></font></font></font></p>
</ul>

<br/>

</p>

<font color="#000000"><font face="FreeSans, sans-serif"><font size="4" style="font-size: 14pt"><span style="background: #ffffff">and
twelve data quality issues&nbsp;that were either visually or
programmatically identified:</span></font></font></font></p>
<ul>
